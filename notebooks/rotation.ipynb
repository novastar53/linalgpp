```python
import numpy as np

# define a rotation function
def rotate(v: np.ndarray, theta: float) -> np.ndarray:
    assert len(v) == 2

    r = np.zeros((2,2))
    r[0,0] = np.cos(theta)
    r[0,1] = -np.sin(theta)
    r[1,0] = np.sin(theta)
    r[1,1] = np.cos(theta)

    v_prime = r @ v
    return v_prime

# specify a vector 
v = np.array([1,2])

# animate the rotation 
v_prime = rotate(v, np.pi/2)

import matplotlib.pyplot as plt

plt.quiver(0, 0, v[0], v[1], scale=10) #, scale_units='xy', scale=1)
plt.quiver(0, 0, v_prime[0], v_prime[1], scale=10, color='red') #, angles='xy', scale_units='xy', scale=1)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
```
